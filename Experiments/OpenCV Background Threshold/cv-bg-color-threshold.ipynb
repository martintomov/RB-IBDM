Using Grayscale

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, measure
```

```python
imgPath = 'data/1.jpg'
img = cv.imread(imgPath)
imgGray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
hist = cv.calcHist([imgGray], [0], None, [256], [0, 256])
plt.figure()
plt.plot(hist)
plt.xlabel('bins')
plt.ylabel('# of pixels')
plt.show()
```

```python
thresOpt = [cv.THRESH_BINARY,
            cv.THRESH_BINARY_INV]

thresNames = ['binary', 'binaryInv']
```

```python
LOW = 135
HIGH = 255

for i in range(len(thresOpt)):
  plt.subplot(2, 3, i+2)
  _, imgThres = cv.threshold(imgGray, LOW, HIGH,thresOpt[i])
  plt.imshow(imgThres, cmap='gray')
  plt.title(thresNames[i])
plt.show
```

KMeans clustering

```python
import matplotlib as mpl
from sklearn.cluster import KMeans
```

```python
image = io.imread('data/4.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
X = image.reshape(-1, 3)
```

```python
X.shape
```

```python
kmeans = KMeans(n_clusters=2, n_init=10)
```

```python
kmeans.fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img / 255)
```
